```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("datasets/cefr_leveled_texts.csv")
```

```python
data.head()
```

```python
unique_level = data["label"].unique()
unique_level
```

```python
class_counts = data["label"].value_counts()
print("Number of Levels:\n", class_counts)
```

```python
qa_format = [
        {
            "messages": [
                {"role": "system", "content": "An assistant that creates level based English texts."},
                {'role': 'user', 'content': f"Generate a text in English demonstrating proficiency at the {row['label']} level."},
                {'role': 'assistant', 'content': f"Here is a text in English at proficiency {row['label']} level\n{row['text']}"},
            ]
        } 
        for index, row in data.iterrows()
    ]

qa_format
```

```python
len(qa_format)
```

```python
from json import dumps

file_path = "./json_files/jsonl_texts.jsonl"
with open(file_path, 'w') as jsonl_file:
    for pair in qa_format:
        jsonl_file.write(dumps(pair) + '\n')

print("jsonl file for texts has been successfuly created!")
```
